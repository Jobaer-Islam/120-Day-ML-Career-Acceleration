# Hint 9.1 – Basic Yield vs Return

Three examples showing how `yield` differs from `return`.

## Example 1: Simple Generator

```python

def gen_numbers():
    yield 1
    yield 2
    yield 3

g = gen_numbers()
print(next(g))
print(next(g))
print(next(g))
```

## Example 2: return ends function

```python

def normal_func():
    return 1
    return 2

print(normal_func())
```

## Example 3: yield pauses execution

```python

def pause_demo():
    print("Start")
    yield "Paused"
    print("Resumed")

g = pause_demo()
print(next(g))
print(next(g))
```

# Hint 9.3 – Infinite Sequences

Using generators for infinite loops safely.

## Example 1: Infinite counter

```python

def counter():
    i = 0
    while True:
        yield i
        i += 1

c = counter()
for _ in range(5):
    print(next(c))
```

## Example 2: Fibonacci sequence

```python

def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

f = fibonacci()
for _ in range(7):
    print(next(f))
```

## Example 3: Controlled infinite loop

```python

def infinite_even():
    n = 0
    while True:
        yield n
        n += 2

g = infinite_even()
for _ in range(5):
    print(next(g))
```

# Hint 9.4 – Generator Exhaustion

Generators can only be iterated once.

## Example 1: Exhausted generator

```python

g = (i for i in range(3))
print(list(g))
print(list(g))  # Empty second time
```

## Example 2: StopIteration manually

```python

def gen():
    yield 1

g = gen()
print(next(g))
try:
    print(next(g))
except StopIteration:
    print("Generator exhausted")
```

## Example 3: Recreate generator

```python

def gen():
    yield 1
    yield 2

g1 = gen()
g2 = gen()
print(list(g1))
print(list(g2))
```

# Hint 9.5 – for-loop & next() Protocol

How for-loops work internally.

## Example 1: Manual iteration

```python

it = iter([1, 2, 3])
while True:
    try:
        print(next(it))
    except StopIteration:
        break
```

## Example 2: for-loop equivalent

```python

for x in [1, 2, 3]:
    print(x)
```

## Example 3: Generator with for-loop

```python

def gen():
    for i in range(3):
        yield i

for x in gen():
    print(x)
```

# Hint 9.6 – Generator Pipelines

Chaining generators together.

## Example 1: Square pipeline

```python

def numbers():
    for i in range(5):
        yield i

def squares(nums):
    for n in nums:
        yield n * n

for val in squares(numbers()):
    print(val)
```

## Example 2: Filter pipeline

```python

def evens(nums):
    for n in nums:
        if n % 2 == 0:
            yield n

for x in evens(range(10)):
    print(x)
```

## Example 3: Multi-stage pipeline

```python

def stage1():
    for i in range(5):
        yield i

def stage2(data):
    for x in data:
        yield x + 10

def stage3(data):
    for x in data:
        yield x * 2

for result in stage3(stage2(stage1())):
    print(result)
```

# Hint 9.7 – Large File Reader

Simulating large file reading.

## Example 1: Line by line reading

```python

def fake_file():
    for i in range(5):
        yield f"Line {i}"

for line in fake_file():
    print(line)
```

## Example 2: Chunk reader

```python

def chunk_reader(data, size):
    for i in range(0, len(data), size):
        yield data[i:i+size]

for chunk in chunk_reader("ABCDEFGHIJK", 3):
    print(chunk)
```

## Example 3: Streaming processing

```python

def reader():
    for i in range(5):
        yield i

def processor(data):
    for x in data:
        yield x * 100

for val in processor(reader()):
    print(val)
```

# Hint 9.8 – yield from

Delegating to sub-generators.

## Example 1: yield from list

```python

def gen():
    yield from [1, 2, 3]

for x in gen():
    print(x)
```

## Example 2: Nested generators

```python

def inner():
    yield 1
    yield 2

def outer():
    yield 0
    yield from inner()
    yield 3

for x in outer():
    print(x)
```

## Example 3: yield from with range

```python

def gen():
    yield from range(5)

print(list(gen()))
```

# Hint 9.9 – send() Method

Sending data into generators.

## Example 1: Basic send

```python

def coroutine():
    x = yield
    print("Received:", x)

c = coroutine()
next(c)
c.send("Hello")
```

## Example 2: Accumulator

```python

def accumulator():
    total = 5
    while True:
        val = yield total
        total += val

acc = accumulator()
next(acc)
print(acc.send(10))
print(acc.send(5))
```

## Example 3: Interactive generator

```python

def echo():
    while True:
        msg = yield
        print("Echo:", msg)

e = echo()
next(e)
e.send("Hi")
e.send("Bye")
```

# Hint 9.10 – State Retention

Generators as stateful functions.

## Example 1: Running total

```python

def running_total():
    total = 0
    while True:
        total += yield total

rt = running_total()
next(rt)
print(rt.send(3))
print(rt.send(7))
```

## Example 2: Stateful counter

```python

def counter():
    n = 0
    while True:
        yield n
        n += 1

c = counter()
for _ in range(4):
    print(next(c))
```

## Example 3: Stateful toggle

```python

def toggle():
    state = False
    while True:
        state = not state
        yield state

t = toggle()
for _ in range(5):
    print(next(t))
```
